```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfA = pd.read_csv('Aluminium_Profile.csv', sep='\t+')
dfB = pd.read_csv('Brass_Profile.csv', sep='\t+')
dfS = pd.read_csv('Steel_Profile.csv', sep='\t+')

dfS.head()
```

```python
plt.figure(figsize=(8,6))

plt.plot(dfA['Radial Modelled (mm)'], dfA[' Height Modelled (mm)'], linestyle='-', color='blue', label='Aluminium')
plt.plot(dfB['Radial Modelled (mm)'], dfB[' Height Modelled (mm)'], linestyle='-', color='magenta', label='Brass')
plt.plot(dfS['Radial Modelled (mm)'], dfS[' Height Modelled (mm)'], linestyle='-', color='green', label='Steel')

r_al = 0.68
r_brass = 0.7
r_steel = 0.62
plt.axvline(x=r_al, color='blue', linestyle='--', linewidth=1)
plt.axvline(x=r_brass, color='magenta', linestyle='--', linewidth=1)
plt.axvline(x=r_steel, color='green', linestyle='--', linewidth=1)
plt.text(r_al + 0.02, -0.09, 'd_Al/2 = 0.68mm', color='blue')
plt.text(r_brass + 0.02, -0.045, 'd_Brass/2 = 0.70mm', color='magenta')
plt.text(r_steel + 0.02, -0.135, 'd_Steel/2 = 0.62mm', color='green')

plt.title('Surface Profiles After Indentation')
plt.xlabel('Radial Distance from Indentation Centre (mm)')
plt.ylabel('Height from Base Plane (mm)')
plt.legend()

plt.savefig('Profile.png', dpi=1000)
plt.show()
```
