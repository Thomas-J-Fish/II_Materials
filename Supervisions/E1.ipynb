```python
import numpy as np
import matplotlib.pyplot as plt

ns = np.linspace(0.001, 100, 10000)
sigma_f = (1 - (np.tanh(ns) / ns))

plt.plot(ns, sigma_f, linestyle='-')
plt.xlabel("ns")
plt.ylabel("sigma_f")
plt.title("Plot of Average Fibre Stress")
plt.show()

sigma_f_max = (1 - (np.tanh(100) / 100))
sigma_f_90 = 0.9 * sigma_f_max

idx = np.argmin(np.abs(sigma_f - sigma_f_90))
print("ns for 90% max =", ns[idx], "sigma_f =", sigma_f[idx])
```
