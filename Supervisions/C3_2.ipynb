```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C3_data2.csv",header=0)
df.head()
```

```python
plt.plot(df["Eph"],df["alpha"])
plt.xlabel("Eph / ev")
plt.ylabel("alpha / m^-1")
plt.title("Tauc plot")
plt.text(2.1,6e+6,"Indirect band structure indicated \n by the convex gradient behaviour \n reflecting the square relationship.")
plt.show()
```
