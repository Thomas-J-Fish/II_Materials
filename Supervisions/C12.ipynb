```python
import numpy as np
import matplotlib.pyplot as plt

k = 52.0
mu_vals = [0,0.1,0.2,0.3]
h = 2.0
b = 16.0
x = np.linspace(0,b/2,200)

stk = 2*k*(1 + (1/h)*(b/2 - x))

plt.figure(figsize=(8,5))
plt.plot(x, stk, label="Sticking Friction", color="black")

for mu in mu_vals:
    slp = 2*k*np.exp((2*mu/h)*(b/2 - x))
    line, = plt.plot(x, slp, label=f"Slipping Friction ($\mu$ = {mu})")
    if mu != 0:
        x_trans = b/2 - (h/(2*mu))*np.log(1/(2*mu))
        if x_trans >=0 and x_trans <= 8:
            plt.axvline(x_trans, color=line.get_color(), linestyle=":")

plt.title("Die Pressure Distribution (Right Half)")
plt.xlabel("Distance from center, x (cm)")
plt.ylabel("Die Pressure, p(x) (MPa)")
plt.legend()
plt.grid(True)

plt.show()
```
